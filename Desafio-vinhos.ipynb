```python
id_pais = dados[["Id","País"]]
id_pais.head()
```

```python
ultimos_15anos = dados[dados.columns[-30:]]
ultimos_15anos.head()
```

```python
ultimos_15anos.head()
```

```python
colunas = ultimos_15anos.columns.str.replace(".", "-Dolar")
colunas 
```

```python
colunas = colunas.str.replace("r1", "r")
```

```python
ultimos_15anos.columns = colunas
ultimos_15anos.head()
```

```python
valores_dolar = []
for i in colunas :
    if str(i).find("Dolar")!= -1 :
        valores_dolar.append(i)
    
```

```python
valores_dolar
```

```python
dados_dolar = id_pais.join(ultimos_15anos[valores_dolar])
```

```python
dados_dolar.head()
```

```python
dados_dolar.describe()
```

```python
dados_dolar["Total"] = dados_dolar.iloc[:,2:].sum(axis = 1)
```
